```python
from scipy.stats import kruskal
import pandas as pd

cluster_col = 'marker_cluster'

# Step 1: Select only clinical phenotype columns (exclude clusters and IDs)
clinical_cols = [
    col for col in df_pheno_marker_merge.columns
    if col != cluster_col and
       df_pheno_marker_merge[col].dtype != 'O' and  # Exclude object (string) types
       col not in ['harm_id', 'pheno-cluster']
]

# Step 2: Run Kruskal-Wallis test for each phenotype
results = []

for feature in clinical_cols:
    temp_df = df_pheno_marker_merge[[cluster_col, feature]].dropna()
    groups = [group[feature].values for _, group in temp_df.groupby(cluster_col)]

    # Only test if there are at least 2 valid groups with more than 1 sample each
    if len(groups) >= 2 and all(len(g) > 1 for g in groups):
        stat, p = kruskal(*groups)
        results.append({'feature': feature, 'H_statistic': stat, 'p_value': p})

# Step 3: Compile and sort results
results_df = pd.DataFrame(results).sort_values('p_value')

# Display top results
print(results_df.head(10))
```

```python
df_labevents = pd.read_csv(URL_DATA+'hosp/labevents.csv.gz')
```

```python
df_timeseries_patient = pd.merge(df_timeseries_patient, df_labevents[['charttime', 'hadm_id']])
```

```python
groups = df_timeseries_patient.groupby('subject_id')
sequences = []
feature_names = None

for subject_id, group in groups:
    sorted_group = group.sort_values('charttime')
    features = sorted_group.drop(['subject_id', 'hadm_id','charttime'], axis=1)
    sequences.append(features)
    if feature_names is None:
        feature_names = features.columns.tolist()
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

padded = pad_sequences(sequences, padding='post', dtype='float32')
```

```python
from ts2vec import TS2Vec
import torch

model = TS2Vec(input_dims=padded.shape[-1], device='cuda',batch_size=16)

MAX_LEN = 512
padded = padded[:, :MAX_LEN, :]
model.fit(padded)

embeddings = model.encode(padded, encoding_window='full_series')
```

```python
import hdbscan

clusterer = hdbscan.HDBSCAN(min_cluster_size=5)
labels = clusterer.fit_predict(embeddings)
```

```python
from umap import UMAP
import matplotlib.pyplot as plt

umap_proj = UMAP(n_components=2, random_state=42).fit_transform(embeddings)

plt.figure(figsize=(8,6))
plt.scatter(umap_proj[:,0], umap_proj[:,1], c=labels, cmap='tab10', s=42)
plt.title('UMAP projection of TS2Vec embeddings colored by HDBSCAN clusters')
plt.show()
```

```python
collapsed = padded.mean(axis=1)  # shape: (n_subjects, n_features)
n_features = collapsed.shape[1]

feature_influence = np.zeros((n_features, embeddings.shape[1]))
for i in range(n_features):
    for j in range(embeddings.shape[1]):
        feature_influence[i,j] = np.corrcoef(collapsed[:,i], embeddings[:,j])[0,1]

# Sum absolute correlations across embedding dimensions to rank features
overall_influence = np.abs(feature_influence).sum(axis=1)
top_features_idx = np.argsort(overall_influence)[::-1]
```

```python
import matplotlib.pyplot as plt

for feature_idx in top_features_idx[:10]:  # top 3 features
    plt.figure(figsize=(8,6))
    for cluster in np.unique(labels):
        cluster_seqs = padded[labels==cluster, :, feature_idx]
        mean_traj = cluster_seqs.mean(axis=0)
        std_traj = cluster_seqs.std(axis=0)
        plt.plot(mean_traj, label=f'Cluster {cluster}')
        plt.fill_between(range(len(mean_traj)), mean_traj-std_traj, mean_traj+std_traj, alpha=0.3)
    plt.title(f'Feature {feature_names[feature_idx]} trajectories by cluster')
    plt.savefig(f'Trajectories/Feature-{feature_names[feature_idx]}-trajectories-by-cluster.png')
    plt.legend()
    plt.show()
```
